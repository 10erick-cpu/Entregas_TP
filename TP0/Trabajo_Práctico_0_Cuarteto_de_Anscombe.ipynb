#**An√°lisis Grafico**


1.   Procedemos a realizar los gr√°ficos de los 4 datasets,  Y = f(x) .
2.   Luego evaluamos los 4 nuevos datos, para saber  que puntos de esos datos se corresponden con cada dataset del cuarteto de Anscombe.




#**Datos nuevos a verificar**

```python
new_values_x = [16,8,16,16]
new_values_y = [8,6,10,6]
```

#**Gr√°ficos**

# **Gr√°fico - Dataset I**

```python
df1 = pd.read_csv("Dataset Cuarteto de Anscombe.csv",usecols = ['X1','Y1'],sep=';')

x = np.array(df1['X1'])
y = np.array(df1['Y1'])


plt.figure(figsize=(10,7.5))
plt.plot(x,y,'go',label='Dataset - I')
plt.plot(new_values_x,new_values_y,'bo',label='Datos a verificar')


plt.grid()
plt.xlabel('X1',fontsize=12)
plt.ylabel('Y1',fontsize=12)
plt.title('Cuarteto de Anscombe - Dataset I',fontsize=15)
plt.legend( loc='lower right',fontsize=15)


# ax = plt.gca()
# ax.set_facecolor('white')

plt.show()

```

# **Conclusi√≥n** :

####Seg√∫n gr√°fico se deduce que el punto (dato nuevo) a verificar :


new_value_1 = (16,8)   pertenece al Dataset I
